One Numeric Variable

Histogram

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
URL = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv"
data = pd.read_csv(URL)
data
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 8))
ax.set_xlim(0,300)
sns.histplot(data.price, bins=1000,ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6, 8))
ax.set_xlim(0,300)
sns.histplot(data.price, bins=3000,ax=ax)
```

Density Plot

```python
fig, ax = plt.subplots(figsize=(6, 8))
sns.kdeplot(data.price, x="price", ax = ax, bw_adjust=1, fill=True, cut =10)
ax.set_xlim(0, 1000)
```

```python
fig, ax = plt.subplots(figsize=(6, 8))
sns.kdeplot(data.price, x="price", ax = ax, bw_adjust=0.05, fill=True, cut =2)
ax.set_xlim(0, 1000)
```
